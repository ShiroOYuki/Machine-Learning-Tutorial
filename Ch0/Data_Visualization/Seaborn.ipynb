```python
import seaborn as sns
import pandas as pd
from sklearn import datasets
```

```python
iris = datasets.load_iris()
print(iris.keys())
df = pd.DataFrame(iris["data"],columns=iris["feature_names"])
target = pd.DataFrame(iris["target"],columns=["target"])
mapping = {
    0:"setosa",
    1:"versicolor",
    2:"virginica"
}
target["target"] = target["target"].map(mapping)
df = pd.concat([df,target],1)

df
```

```python
l = [
    [1,5],
    [2,-6],
    [3,1],
    [4,2],
    [5,-3],
    [6,8],
    [7,1],
    [8,-9]
]
column = ["x","y"]

df2 = pd.DataFrame(l,columns=column)
df2
```

# 折線圖 sns.lineplot()
* `sns.lineplot(x=<x lable:str>, y=<y lable:str>, data=<data:pandas.DataFrame>, hue=[column name])`
    
    > x：x 軸名稱  
    > y：y 軸名稱  
    > data：資料內容 ( 必須是 DataFrame )  
    > hue：顏色 ( 根據填入的欄位分類 )  

```python
g = sns.lineplot(x="x",y="y",data=df2)
```

```python
g = sns.lineplot(x="sepal length (cm)",y="sepal width (cm)",data=df,hue="target")
```

# 長條圖 sns.barplot()
* `sns.barplot(x=<x lable:str>, y=<y lable:str>, data=<data:pandas.DataFrame>)`
    
    > x：x 軸名稱  
    > y：y 軸名稱  
    > data：資料內容 ( 必須是 DataFrame )  

```python
g = sns.barplot(x="x",y="y",data=df2)
```

# 散布圖 sns.lmplot()
* `sns.lmplot(x=<x lable:str>, y=<y lable:str>, data=<data:pandas.DataFrame>, hue=[column name], fit_reg=[bool])`
    
    > x：x 軸名稱  
    > y：y 軸名稱  
    > data：資料內容 ( 必須是 DataFrame )  
    > hue：顏色 ( 根據填入的欄位分類 )  
    > fit_reg：是否畫出直線  

```python
g = sns.lmplot(x="sepal length (cm)",y="sepal width (cm)",data=df,hue="target")
g = sns.lmplot(x="sepal length (cm)",y="sepal width (cm)",data=df,hue="target",fit_reg=False)
```
